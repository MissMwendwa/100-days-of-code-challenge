# Matplotlib visualization

We are represnting some simple graphs in MatplotLib of the sine, cosine and tangent mathematical formulae.

```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)
```

```python
plt.plot(x, y, color="red")
plt.show()
```

```python
sin = np.linspace(0, 20, 100)
var = np.sin(sin)
```

```python
plt.plot(sin, var, color = "green")
plt.show()
```

```python
# this is a graph of y = x**2
arr = np.array([12, 14, 16, 18, 20, 24])
arr_2 = (arr)**2
```

```python
plt.plot(arr, arr_2, color = "blue")
plt.show()
```
